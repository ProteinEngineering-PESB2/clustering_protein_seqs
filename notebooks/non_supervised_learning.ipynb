```python
import sys
sys.path.insert(0, '../src')
import pandas as pd
from non_supervised_learning.algorithms import Algorithm
from non_supervised_learning.metrics import Performance
```

## Numerical Representation

```python
from numerical_representation.physicochemical_properties import Physicochemical
```

```python
df = pd.read_csv("../data/absortion.csv")
physicochemical = Physicochemical(df, "Group_0", "../input_encoders/clustering_encoders.csv", "sequence", "id")
df_physicochemical = physicochemical.encode_dataset()
df_physicochemical
```

```python
df_physicochemical = df_physicochemical.drop(["id"], axis =1)
```

## Kmeans

```python
algorithm = Algorithm()
model = algorithm.apply_kmeans(df_physicochemical)
performance = Performance()
metrics = performance.get_performance(df_physicochemical, model)
metrics
```

```python
kmeans = df_physicochemical.copy()
kmeans["id"] = df["id"]
if metrics is not None:
    kmeans["cluster"] = model.labels_
else:
    kmeans["cluster"] = None
kmeans
```
